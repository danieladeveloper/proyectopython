```python
import pandas as pd #pandas para la manipulación de datos
import matplotlib.pyplot as plt #para la visualización de datos

data_path="../data/processed/cleaned_products.csv"
df = pd.read_csv(data_path) #cargamos los datos del archivo

print("Basic Statistics")
print(df.describe()) #muestra el resumen estadístico de los datos en el df

#histograma

plt.hist(df["price"], bins=10, edgecolor="black") #creamos un histograma de "Price con 20 barras y bordes negros"
plt.title("Distribución de precios") #título del histograma
plt.xlabel("Precio") #etiquete eje x
plt.ylabel("Frecuencia") # etiqueta eje y
plt.show() #mostrar el Histrograma

print("Productos más caros: ")
highestPrice= df.nlargest(5,"price")
print(highestPrice)
```
